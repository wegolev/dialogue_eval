### Задача 2. Определение тональности аспектных терминов.
По отношению к каждому из выделенных терминов необходимо указать его тональность по 4-х бальной шкале:
    • позитивная;
    • негативная;
    • противоречивая (по отношению к термину выражены как позитивная, так и негативная тональность);
    • нейтральная (по отношению к термину не выражено позитивной или негативной тональности).
Например, во фрагменте «кухня хорошая, однако интерьер слишком уж вычурный и броский» аспектный термин «кухня» имеет позитивную тональность, а термин «интерьер» негативную тональность.
Во фрагменте «Обслуживание может и не самое лучшее в городе, но официанты делали всё что нужно» термин «обслуживание» имеет противоречивую тональность.
Во фрагменте «поскольку настроение у всех было отличное и расходиться не хотелось, мы заказали чай и кофе» термины «чай» и «кофе» имеют нейтральную тональность.

### Задача 3. Определение целевого аспекта термина.
Каждый выделенный термин в отзывах о ресторанах необходимо отнести к одной из пяти аспектных категорий:
    • кухня;
    • интерьер;
    • сервис;
    • цена;
    • ресторан в целом.
Например, «Довольно хорошие впечатления от этого ресторана. Остался в восторге от салата с говядиной, и конечно стейк из говядины! Еще не могу не отметить крем-суп из лосося! Да, обслуживающая девушка некоторые моменты забывала, приходилось напоминать. Про интерьер – действительно, атмосфера теплая, мягкая и расслабляющая. Довольно удобные диваны и кресла».
    • Термин «ресторан» относится к аспекту «ресторан в целом».
    • Термины «салата с говядиной», «стейк из говядины», «крем-суп из лосося» определяют аспект «кухня».
    • Термин «обслуживающая девушка» относится к аспекту «сервис».
    • Термины «атмосфера», «диваны» и «кресла» относятся к аспекту «интерьер».
Каждый выделенный термин в отзывах об автомобилях необходимо отнести к одной из следующих аспектных категорий:
    • комфорт,
    • надежность, 
    • внешний вид, 
    • цены, 
    • дорожные качества,
    • автомобиль в целом (whole).

### Задача 4. Определение интегральных оценок.
Для всего отзыва нужно определить результирующие тональности (по той же шкале, что и тональность терминов) для каждого из целевых аспектов по шкале:
    • позитивная;
    • негативная;
    • противоречивая;
    • нейтральная (термины категории присутствуют, но оценка не выражена).
    • Если в тексте отзыва нет упоминаний терминов категории, то оценка не проставляется.

Например, для примера выше интегральные оценки по аспектам будут следующими:
    • кухня – позитивная;
    • интерьер – позитивная;
    • сервис – противоречивая;
    • ресторан в целом – позитивная.

Оценка объекта в целом в большинстве случаев должна проставляться как вывод по отзыву в целом (позитивный или негативный): рекомендуют ли приходить в ресторан или нет, рекомендуют ли покупать автомобиль (хотя бы каким-то категориям покупателей) или нет.
